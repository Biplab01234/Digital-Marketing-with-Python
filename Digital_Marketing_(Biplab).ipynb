Q12.	Are there any outliers in terms of cost, clicks, or revenue that warrant further investigation?

```python

# Define a function to find outliers using IQR
def find_outliers_iqr(data, column_name):
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    return outliers

# Find outliers for cost, clicks, and revenue
cost_outliers = find_outliers_iqr(df, 'cost')
clicks_outliers = find_outliers_iqr(df, 'clicks')
revenue_outliers = find_outliers_iqr(df, 'revenue')

# Display the results
print("Outliers in cost:")
print(cost_outliers[['cost', 'clicks', 'revenue']])

print("\nOutliers in clicks:")
print(clicks_outliers[['cost', 'clicks', 'revenue']])

print("\nOutliers in revenue:")
print(revenue_outliers[['cost', 'clicks', 'revenue']])
```
